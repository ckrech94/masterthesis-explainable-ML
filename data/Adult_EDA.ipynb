## Explorative Data Analysis Adult income dataset  



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pl
import seaborn as sns

from dataset import adult

import warnings
warnings.filterwarnings('ignore')
```

### Laden des Datensatzes

```python
X,y = adult()
X_display, y_display = adult(display=True)
```

### Verteilungsplots der Merkmale

```python
plt.figure(figsize=(6,5), facecolor=None, edgecolor=None)
sns.set_style("white")
ax = sns.distplot( X['Age'],bins=24,color = "#1E88E5",kde=False )
ax.set_xlabel('Age', color="#333333", fontsize=13)
ax.set_ylabel('Absolute Häufigkeit \n von Age', color="#333333", fontsize=13)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```
